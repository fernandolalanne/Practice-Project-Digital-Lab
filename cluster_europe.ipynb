# economics

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import plotly.graph_objs as go
import plotly.express as px
import dash
from dash import html
from dash import dcc


df_economics_europe = pd.read_csv('df_economics_europe.csv', delimiter=',')

df_clustering = df_economics_europe.select_dtypes(include=['float64', 'int64'])

df_clustering = df_clustering.fillna(df_clustering.mean())

scaler = StandardScaler()
df_clustering_scaled = scaler.fit_transform(df_clustering)

# inertia = []
# for i in range(1, 11):
#     kmeans = KMeans(n_clusters=i, random_state=42)
#     kmeans.fit(df_clustering_scaled)
#     inertia.append(kmeans.inertia_)

# silhouette_scores = []
# for i in range(2, 11): # Silhouette score no se puede calcular para un solo cluster
#     kmeans = KMeans(n_clusters=i, random_state=42)
#     kmeans.fit(df_clustering_scaled)
#     score = silhouette_scores(df_clustering_scaled, kmeans.labels_)
#     silhouette_scores.append(score)c


kmeans_final = KMeans(n_clusters=4, random_state=42) # Cambia 3 por el número óptimo encontrado
kmeans_final.fit(df_clustering_scaled)
df_economics_europe['Cluster'] = kmeans_final.labels_



# plt.figure(figsize=(10, 6))
# plt.plot(range(1, 11), inertia, marker='o', linestyle='-', color='blue')
# plt.title('Método del Codo para Determinar el Número Óptimo de Clusters')
# plt.xlabel('Número de Clusters')
# plt.ylabel('Inercia')
# plt.grid(True)
# plt.show()


# plt.figure(figsize=(10, 6))
# plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='-', color='red')
# plt.title('Silhouette Score para Determinar el Número Óptimo de Clusters')
# plt.xlabel('Número de Clusters')
# plt.ylabel('Silhouette Score')
# plt.grid(True)
# plt.show()
```

```python
moda_clusters = df_economics_europe.groupby('location')['Cluster'].agg(pd.Series.mode)

clusters_moda_dict = moda_clusters.to_dict()

for key, value in clusters_moda_dict.items():
    if isinstance(value, pd.Series):
        clusters_moda_dict[key] = value.iloc[0]

print(clusters_moda_dict)
```

```python
df_economics_europe['Cluster2'] = df_economics_europe['location'].map(clusters_moda_dict)
```
